```python
a = int(input("the number 1 is : "))
b = int(input("the number 2 is : "))

def countOne(n) :
  count = 0
  while(n) :
        count += 1
        n &= (n - 1)
  return count

def countFlips(a , b) :
  return countOne(a ^ b)        
print(countFlips(a , b))
```
